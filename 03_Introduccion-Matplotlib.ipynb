## Reprecentacion grafica de datos 
si la funcion de trazados se le da una matriz de datos , la usara como coordenadas en el eje vertical , utilizara el indice de cada punto de datos en el array como la coordenada horizontal 

```python
plt.plot([1, 2, 3, 4, 5, 7, 8, 3, 1])
plt.show()
```

Tambien se puede proporcionar dos matrices : una para el eje horizontal y otra para el eje vertical  

```python
plt.plot([-3, -1,  0, 4, 7], [1, 4, 6, 7, 8])
plt.show()
```

pueden modificarse  las longitudes de los ejes para que la grafica no se vea tan ajustada 

```python
plt.plot([-3, -1,  0, 4, 7], [1, 4, 6, 7, 8])
plt.axis([-4, 8, 0, 10])            #[Xmin, xmax,ymin,ymax]
plt.show
```

se sigue el mismo procediminto para pintar una funcion matematica 

```python
import numpy as np
x = np.linspace (-2, 2, 500)
y = x**2
plt.plot(x, y)
plt.show
```

tambien puede modificarse el estilo de la grafica para que contenga mas informacion 

```python
plt.plot(x, y)
plt.title("Funcion Cuadratica ")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show
```

pueden superponerse graficas y cambiar el estilo de las lineas 

```python
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
y2 = x + 1
plt.plot(x, y,  'b--', x, y2, 'g')
plt.show()
```

para poder diferenciar entre ambas funciones siempre es recomendado añadir una leyenda 


```python
import numpy as np 
x = np.linspace (-2, 2, 500)
y = x**2
y2 = x + 1

plt.plot(x, y,'b--', label = "x**2")
plt.plot(x, y, 'g', label ="x + 1")
plt.legend(loc ="best")
plt.show()

```

para que las graficas no queden tan ajustadas se puede hacer la figura mas grande 

```python
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)     # 1 Rows, 2 columns, 1sl Subplot.
plt.plot (x, y, 'b--')
plt.xlabel("x1", fontsize =14)
plt.xlabel("y2",fontsize =14)

plt.figure(figsize = (14, 6))

plt.subplot(1, 2, 2)     # 1 Rows, 2 columns, 1sl Subplot.
plt.plot (x, y2, 'g')
plt.xlabel("x1", fontsize =14)
plt.xlabel("y2",fontsize =14)
```

## Histogramas 

```python
data =[1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 3]
plt.subplot(211) # 2 Rows, 1 Colunms, 1st Subplot
plt.hist(data, bins = 10, rwidth = 0.8)

plt.subplot(211)   # 2 Rows, 1 Colunms, 1st Subplot
plt.hist(data, bins =[1, 1.5, 2, 2.5, 3], rwidth = 0.95)
plt.xlabel("valor")
plt.ylabel("frecuencia")
plt.show()
```
